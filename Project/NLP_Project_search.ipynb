```
import json
import pymorphy2

morph = pymorphy2.MorphAnalyzer()

with open('corpORAORA.txt') as json_file:
    end_s = json.load(json_file)
```

```
tags = ['INFN', 'ADJS', 'ADJF', 'NOUN', 'PREP']

def is_suitable(pattern, some, place):
    if pattern in tags:
        if pattern in some[4][place]:
            return True
    elif pattern[0] == '"':
        pattern = pattern[1:-1]
        if pattern == some[2][place]:
            return True
    else:
        normie = morph.parse(pattern)[0].normal_form
        if normie == some[3][place]:
            return True
    return False
```

```
def search(some):
    tasks = some.split()
    ans = []
    for some in end_s:
        for i in range(len(some[3]) - len(tasks) + 1):
            u = True
            for j in range(len(tasks)):
                if not (is_suitable(tasks[j], some, i+j)):
                    u = False
                    break
            if u:
                ans.append(some)
                break
                
    return ans
```

```
def beautiful(ans):
    res = []
    for some in ans:
        res.append('Source: ' + str(some[0]) + '; Sentence: ' + str(some[1]))
    return '\n'.join(res)
```

```
query = input()

result = beautiful(search(query))
print(result)
```
